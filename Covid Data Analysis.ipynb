```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
covid_data = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
covid_data
```

```python
covid_data.info()
```

```python
covid_data.describe()
```

```python
covid_data.head()
```

```python
total_states = np.arange(len(covid_data['state_name']))
total_states #number of states
```

```python
covid_data['active'] #number of active cases
```

```python
max(covid_data
    ['positive'])
```

# No of positive cases in each state

```python
plt.figure(figsize =(11,9))

plt.barh(total_states,covid_data['positive'], align='center', alpha=0.5,color=(1,0,0),  edgecolor=(0.5,0.2,0.8))
    
plt.yticks(total_states, covid_data['state_name'])  
plt.xlim(1,max(covid_data['positive'])+100) 
plt.xlabel('Number of Positive Cases')  
plt.title('No of COVID Cases')  
plt.show()  
```

# Active cases in each state

```python
plt.figure(figsize =(11,9))

plt.barh(total_states,covid_data['new_active'], align='center', alpha=0.5,  color=(1,0,1),  
                 edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(total_states, covid_data['state_name'])  
plt.xlim(1,max(covid_data['new_active'])+10) 
plt.xlabel('Number of active states')  
plt.title('No of COVID Cases')  
plt.show()  
```

# Number of deaths 

```python
plt.figure(figsize =(11,9),dpi =80)

plt.barh(total_states,covid_data['death'], align='center', alpha=0.5,  color=(0,0,1),  edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(total_states, covid_data['state_name'])  
plt.xlim(1,max(covid_data['death'])+10) 
plt.xlabel('Number of death Cases')  
plt.title('No of COVID Cases')  
plt.show()  
```
